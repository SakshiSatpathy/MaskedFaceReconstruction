```python
import requests
from bs4 import BeautifulSoup

# Function to fetch image URLs from Bing
def fetch_bing_image_urls(query, max_links=2800):
    search_url = f"https://www.bing.com/images/search?q={query.replace(' ', '+')}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    # Sending request to Bing Images search page
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all image tags
    img_tags = soup.find_all("img", {"class": "mimg"})
    img_urls = []

    # Collect image URLs
    for img in img_tags:
        img_url = img.get("src")
        if img_url and len(img_urls) < max_links:
            img_urls.append(img_url)

    return img_urls

# Download images from URLs
def download_images(urls, label):
    import os
    import urllib.request

    # Create folder if it does not exist
    folder_path = f"./{label}"
    os.makedirs(folder_path, exist_ok=True)

    for idx, url in enumerate(urls):
        try:
            # Download each image
            image_path = os.path.join(folder_path, f"{label}_{idx+1}.jpg")
            urllib.request.urlretrieve(url, image_path)
            print(f"Saved {image_path}")
        except Exception as e:
            print(f"Failed to download {url}: {str(e)}")

queries = [
    "White person face", "Black person face", "Asian person face",
    "Indian person face", "Latino person face", "Native American person face", "Middle Eastern person face",
    "White person wearing face mask", "Black person wearing face mask", "Asian person wearing face mask",
    "Indian person wearing face mask", "Latino person wearing face mask", "Native American person wearing face mask", "Middle Eastern person wearing face mask"
]

# Iterate through queries and fetch images
for query in queries:
    print(f"\nFetching: {query}")
    label = query.replace(" ", "_")
    image_urls = fetch_bing_image_urls(query, max_links=2800)
    download_images(image_urls, label)
    print(f"Saved {len(image_urls)} images for: {query}")
```
